## Types of numbers; assigning values to variables; the print statement

```python
# Any text after a hash sign is not executed when the program is run.
# This text is called a comment.
# <-- Run this cell by selecting the arrow button at top left.

# Use the equal sign to assign a value to a variable.
# a is an integer

a = 3

# b is a floating point number

b = 1.138

# c and d are complex numbers

c = 1.0 + 3.0j
d = complex(1.0, 3.0)

# e uses exponential notation when assigning a value

e = 2.998e8

# The print statement displays to the terminal its arguments.

print('a =', a)
print('b =', b)
print('c =', c)
print('d =', d)
print('e =', e)
```

## Basic arithmetic operations

```python
# The basic arithmetic operations are:
# +     addition
# -     subtraction
# *     multiplication
# /     floating point division
# //    integer division
# %     modulus
# **    exponentiation

print(3+5)
print(3-5)
a = 3+5
print('a =', a)
print(3 * 5)
print(3 / 5)
print(3 // 5)
print(3 % 5)
print(3**5)
```

## Math functions

```python
# We must import a math library to perform math functions

import math

a = math.sqrt(2)
print(a)
a = math.exp(2)
print(a)
a = math.log(2)
print(a)
a = math.log10(2)
print(a)
a = math.pi
print(a)
a = math.sin(math.pi)
print(a)
a = math.cos(math.pi)
print(a)
a = math.degrees(math.pi)
print(a)
a = math.radians(57.295779513)
print(a)
```

## Comparison operators; boolean logic

```python
# The comparison operators are
# ==     equal to
# !=     not equal to
# >      greater than
# <      less than
# >=     greater than or equal to
# <=     less than ot equal to
#
# The result of a comparison is a boolean object.
# Boolean objects have the value True or the value False.
# To use a boolean operator is to perform an operation and to return
# a value.
#
# The if statement will execute a command only when the if-condition is true.

# In the following line, the variable a is of boolean type.
# It is NOT the letters T, r, u, and e.

a = True
print(a)

print(7 == 8)

# The if ... else structure will execute code conditionally.

if 7 == 8:
  print('This line will not print because the statement is false')
else:
  print('This statement will print')

# You string together boolean operators using and, or, and not.

if 7 == 7 and 8 == 8:
  print('This statement will print because two Trues make a True')
else:
  print('This statement will not print')

if True:
  print('This statement will always print')

if False:
  print('This statement will not print')
```

## Strings

```python
# A string object is an alphanumeric sequence.

string1 = 'abc'
string2 = 'def'

print(string1)
print(string2)

# Strings can be concatenated.

string_total = string1 + string2
print(string_total)

# Strings are indexed starting at zero for the first character.

print(string_total[0])
print(string_total[1])
print(string_total[3])

# So-called slicing of strings selects subsets.
# This line selects index 1, 2, and 3, but not 4.

print(string_total[1:4])
```

## Lists

```python
# Lists are variables with more than value, like vectors and matrices.
# Lists are enclosed in square brackets.

a = [1, 2, 3, 4, 5]
print(a)

# Lists are indexed: the first element has an index of zero.

print(a[1:4])

# Lists can be operated upon.  To append a single value to a list:

a.append(6)
print(a)

# To extend the list with another list:

b = [7, 8, 9]
a.extend(b)
print(a)

# To sort a list:

c = [3, 4, 4, 2 ,6 ,7 ,4, 5 ,6, 5, 1, 9, 8]
print(c)
c.sort()
print(c)

# To count the number of times a 4 appears:

print('Number of times a 4 appears is ',c.count(4))

# Reverse the list

c.reverse()
print(c)
```

## The for loop

```python
# Looping is performing lines of code more than once.
# Lists are iterable. This is useful for pulling items out of a list
# and doing something to each item one at a time.
# Use the for loop.

list = ['orange', 'banana', 'papaya', 'pineapple']

for fruit in list:
  print(fruit)

# Looping over a range of numbers uses the range command.
# The range command creates a list of integers.
# This command loops from 1 to 4, but not 5.

for i in range(1,5):
  print(i)

# Loop from 1 to 9 and stepping by 2.

for i in range(1,10,2):
  print(i)

# The range command produces integers only.
# For non-integer steps, use something like this:

for i in range(1,10):
  print(i/10)

# ... or use the numpy library and the arange command.
# Beware of cumulative rounding errors.

import numpy as np

for i in np.arange(16, 17, 0.2):
  print(i)
```

## The while loop

```python
# The while loop will execute a code block until a condition is met.

i = 0

while i < 5:
  print(i)
  i += 1
```

## Plotting data: scatter plots

```python
# The Python library Matplotlib.pyplot has a comprehensive set of plotting tools.
# The plt.scatter() function makes the plot from x1 and y1 lists.

import matplotlib.pyplot as plt

x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.scatter(x1, y1, color='green', marker='o')
plt.show()
```

## Plotting functions: line plots

```python
# The plt.plot() function connects each (x,y) value with a line.

# Start by defining a pair of empty lists so we can append values to them.

x2 = []
y2 = []

for i in range(10,51):
  x2.append(i/10)
  y2.append(i*i/10/10)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.plot(x2, y2)
plt.show()
```

## Combined plots

```python
## You can combine plots, even of different types, into a single pane.

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.scatter(x1,y1, color='green', marker='o')
ax.plot(x2, y2)
plt.show()
```
